```python
import pandas as pd
import numpy as np
df=pd.read_csv("../data/spot_5min_clean.csv",parse_dates=True,index_col=0)
df=df.copy()
print(type(df.index))
print(df.head())
```

```python
for col in ["open", "high", "low", "close", "volume"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")
print(df.dtypes)
```

```python
df["ret_1"]=df["close"].pct_change(fill_method=None)
df["ret_5"]=df["close"].pct_change(5,fill_method=None)
print(df[["close","ret_1","ret_5"]].head(10))
```

```python
df["ema_20"]=df["close"].ewm(span=20,adjust=False).mean()
df["ema_50"]=df["close"].ewm(span=50,adjust=False).mean()
df["ema_diff"]=df["ema_20"]-df["ema_50"]
print(df[["close","ema_20","ema_50","ema_diff"]].tail(10))
```

```python
df["volatility_20"]=df["ret_1"].rolling(window=20).std()
print(df[["ret_1","volatility_20"]].tail(20))
```

```python
df["volatility_100"]=df["ret_1"].rolling(window=100).std()
print(df[["volatility_20","volatility_100"]].tail(20))
```

```python
df["hl_range"]=(df["high"]-df["low"])/df["close"]
print(df[["high","low","close","hl_range"]].tail(10))
```

```python
vol_mean = df["volume"].rolling(50).mean()
vol_std = df["volume"].rolling(50).std()

df["volume_z"] = (df["volume"] - vol_mean) / vol_std

print(df[["volume", "volume_z"]].tail(20))
```

```python
df["sentiment_pressure"]=df["hl_range"]*df["volume_z"]
print(df[["hl_range","volume_z","sentiment_pressure"]].tail(20))
```

```python
df["future_price"]=df["close"]*[1+0.0001]
df["basis"]=(df["future_price"]-df["close"])/df["close"]
print(df[["close","future_price","basis"]].tail(10))
```

```python
print("Rows before dropna:", len(df))
df = df.dropna()
print("Rows after dropna:", len(df))
```

```python
df.to_csv("../data/spot_5min_features.csv")
print("Saved ../data/spot_5min_features.csv")
```

```python
df["delta_proxy"] = df["ema_diff"]
print(df[["ema_diff", "delta_proxy"]].tail(20))
```

```python
df["gamma_proxy"] = df["ema_diff"].diff()
print(df[["ema_diff", "gamma_proxy"]].tail(20))
```

```python
df["vega_proxy"] = df["volatility_20"] - df["volatility_100"]
print(df[["volatility_20", "volatility_100", "vega_proxy"]].tail(20))
```

```python
df["theta_proxy"] = -df["volatility_20"]
print(df[["volatility_20", "theta_proxy"]].tail(20))
```

```python
print("Rows before dropna:", len(df))
df = df.dropna()
print("Rows after dropna:", len(df))
df.to_csv("../data/spot_5min_features.csv", index=False)
print("Saved final spot_5min_features.csv")
```
