```python
import pandas as pd
df = pd.read_csv("../data/raw_spot_5min.csv")
print(df.head())
print(df.tail())
print("Rows:", len(df))
print("Columns:", df.columns.tolist())
```

```python
df=df.drop(columns=["index"])
print(df.columns.tolist())
```

```python
df["Datetime"] = pd.to_datetime(df["Datetime"])
df = df.set_index("Datetime")
df = df.sort_index()

print(df.head())
print(df.tail())
```

```python
df=df.rename(columns={
    "Open":"open",
    "High":"high",
    "Low":"low",
    "Close":"close",
    "Volume":"volume"
    })
print(df.columns.to_list())
```

```python
df=df[["open","high","low","close","volume"]]
```

```python
print("Nans:\n",df.isna().sum())
print("Duplicate Timestamps:",df.index.duplicated())
```

```python
df.to_csv("../data/spot_5min_clean.csv")
print("Saved ../data/spot_5min_clean.csv")
```
